```python
import string
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from numba import jit
```

```python
# Download the set of stopwords from NLTK
nltk.download('punkt')
nltk.download('stopwords')

# @jit(nopython=True)
def clean_text_data(texts):
    stop_words = set(stopwords.words('english'))
    
    for text in texts:
        # Check characters to see if they are in punctuation
        nopunc = [char for char in texts if char not in string.punctuation]
    
        # Join the characters again to form the string.
        nopunc = ''.join(nopunc)
    

    return " ".join([word for word in nopunc.split() if word.lower() not in stop_words])
```

```python
# yelp_data_copy = yelp_data.copy()
# yelp_data_copy['text'] = yelp_data_copy['text'].apply(clean_text_data)

yelp_data.loc[:, 'text'] = yelp_data['text'].apply(clean_text_data)
```

```python
def matrix_factorization(R, P, Q, steps=25, gamma=0.001,lamda=0.02):
    for step in range(steps):
        for i in R.index:
            for j in R.columns:
                if R.loc[i,j]>0:
                    eij=R.loc[i,j]-np.dot(P.loc[i],Q.loc[j])
                    P.loc[i]=P.loc[i]+gamma*(eij*Q.loc[j]-lamda*P.loc[i])
                    Q.loc[j]=Q.loc[j]+gamma*(eij*P.loc[i]-lamda*Q.loc[j])
        e=0
        for i in R.index:
            for j in R.columns:
                if R.loc[i,j]>0:
                    e= e + pow(R.loc[i,j]-np.dot(P.loc[i],Q.loc[j]),2)+lamda*(pow(np.linalg.norm(P.loc[i]),2)+pow(np.linalg.norm(Q.loc[j]),2))
        if e<0.001:
            break
        
    return P,Q
```
